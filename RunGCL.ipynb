```python
#Inference dataset 1
import os
import torch
import numpy as np
from torchvision import transforms
from PIL import Image
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
from warmup.models import fcn_resnet50

# single inference
def infer(model, image_path, device, infer_type):
    model.eval()  # Set model ke mode evaluasi
    with torch.no_grad():
        # Load dan preprocess gambar
        image = Image.open(image_path).convert("RGB")
        preprocess = transforms.Compose([
            transforms.Resize((224, 224)),  # Sesuaikan ukuran gambar sesuai dengan model
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalisasi sesuai model
        ])
        input_tensor = preprocess(image).unsqueeze(0).to(device)  # Tambahkan dimensi batch dan pindah ke device

        # Lakukan inferensi
        output = model(input_tensor)

        if infer_type == 'warmup':
            probabilities = torch.softmax(output, dim=1)  # Mengubah output menjadi probabilitas
            _, predicted = torch.max(probabilities, 1)  # Ambil kelas dengan probabilitas tertinggi
        else:
            predicted = torch.argmax(output.mean(dim=[2, 3]), dim=1)  # Menggunakan rata-rata spatial untuk tipe lain

        return predicted.item()  # Kembalikan label kelas

# Fungsi untuk otomasi testing
def automate_testing(model_path, image_dir, num_classes, device, infer_type):
    """
    Automates the testing process for a classification model.

    Args:
        model_path (str): Path to the trained model file.
        image_dir (str): Directory containing test images.
        num_classes (int): Number of classes in the classification problem.
        device (torch.device): Device to run the model on (CPU or GPU).
    """
    try:
        # Load the model
        model = torch.load(model_path, map_location=device)
        model.to(device)  # Memindahkan model ke device
        model.eval()

        # Initialize lists to store predictions and true labels
        predicted_labels = []
        true_labels = []

        # Iterate through the test images
        for cls in os.listdir(image_dir):
            cls_path = os.path.join(image_dir, cls)
            if not os.path.isdir(cls_path):
                continue
            for img in os.listdir(cls_path):
                image_path = os.path.join(cls_path, img)
                if image_path.endswith(('.png', '.jpg', '.jpeg', '.bmp')):
                    predicted_label = infer(model, image_path, device, infer_type)
                    predicted_labels.append(predicted_label)
                    true_labels.append(int(cls))

        # Create the confusion matrix
        cm = confusion_matrix(true_labels, predicted_labels, labels=list(range(num_classes)))

        # Plot the confusion matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                    xticklabels=list(range(num_classes)),
                    yticklabels=list(range(num_classes)))
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.title("Confusion Matrix")
        plt.show()

        # Compute and print metrics
        accuracy = accuracy_score(true_labels, predicted_labels)
        precision = precision_score(true_labels, predicted_labels, average='weighted')
        recall = recall_score(true_labels, predicted_labels, average='weighted')
        f1 = f1_score(true_labels, predicted_labels, average='weighted')

        print("Accuracy: {:.2f}%".format(accuracy * 100))
        print("Precision: {:.2f}%".format(precision * 100))
        print("Recall: {:.2f}%".format(recall * 100))
        print("F1 Score: {:.2f}%".format(f1 * 100))

        # Generate and print classification report
        target_names = [str(i) for i in range(num_classes)]
        print("\nClassification Report:")
        print(classification_report(true_labels, predicted_labels, target_names=target_names))

    except FileNotFoundError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example Usage
infer_type = 'warmup'

# Define model
if infer_type == 'warmup':
    model = fcn_resnet50(num_classes=3)
    weight_dir = '/workspace/yogiek/gc2/warmup/ckps/H-D/BUS_val/best.ptl'  # Warmup weight
else:
    from main.models import fcn_resnet50, fcn_3x64_gctx
    FT = torch.load('/workspace/yogiek/gc2/warmup/ckps/H-D/BUS_val/best.ptl', map_location='cpu')
    for parameter in FT.parameters():
        parameter.requires_grad = False
    model = {
        "FT": FT,
        "FS": fcn_resnet50(),
        "GS": fcn_3x64_gctx(),
    }
    weight_dir = '/workspace/yogiek/gc2/warmup/ckps/H-D/BUS_val/best-all.ptl'  # OOD weight

# Test dataset 1
image_dir = '/workspace/yogiek/gc2/datasets/H-D/test/1'  # Replace with your images directory
num_classes = 3  # Replace with the number of class in your model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Dataset 1")
automate_testing(weight_dir, image_dir, num_classes, device, infer_type)
```
